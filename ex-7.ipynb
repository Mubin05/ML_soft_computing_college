```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
A = np.random.rand(20)
B = np.random.rand(20)
```

```python
# COG = sum(i * membership_value) / sum(membership_value)
def center_of_gravity(set_values):
    indices = np.arange(len(set_values))  # indices as 0, 1, 2, ..., 19
    return np.sum(indices * set_values) / np.sum(set_values)
# WAV = sum(membership_value) / number_of_elements
def weighted_average_value(set_values):
    return np.sum(set_values) / len(set_values)
```

```python
# Calculate COG and WAV for both sets A and B
cog_A = center_of_gravity(A)
cog_B = center_of_gravity(B)
wav_A = weighted_average_value(A)
wav_B = weighted_average_value(B)

cog_A, cog_B, wav_A, wav_B
```

```python

# --- Plot 1: Center of Gravity (COG) ---
plt.figure(figsize=(3,4))
plt.plot( A, label='Fuzzy Set A', color='blue')
plt.plot( B, label='Fuzzy Set B', color='green')
# Mark COG with vertical lines
plt.plot(cog_A,'ro', color='red', label='COG of A')
plt.plot(cog_B,'ro', color='darkgreen',  label='COG of B')

plt.xlabel('Element Index')
plt.ylabel('Membership Value')
plt.title('Fuzzy Sets A and B with Center of Gravity (COG)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(3,4))
plt.plot( A, label='Fuzzy Set A', color='blue')
plt.plot( B, label='Fuzzy Set B', color='green')

# Mark WAV with horizontal lines
plt.plot(wav_A,'ro', color='red',  label='WAV of A')
plt.plot(wav_B,'s', color='darkgreen', label='WAV of B')

plt.xlabel('Element Index')
plt.ylabel('Membership Value')
plt.title('Fuzzy Sets A and B with Weighted Average Value (WAV)')
plt.legend()
plt.grid(True)
plt.show()
```
