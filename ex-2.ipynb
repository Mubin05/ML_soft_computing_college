```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
```

```python
matrix = np.random.random((100, 100))
```

```python
matrix = np.random.random((100, 100))
means = np.mean(matrix, axis=0)         # Mean of each column
variances = np.var(matrix, axis=0)      # Variance of each column
kurtoses = kurtosis(matrix, axis=0)     # Kurtosis of each column
skewnesses = skew(matrix, axis=0)       # Skewness of each column
```

```python
print("Mean = ")
print(means)
print()
print("Variens = ")
print(variances)
print()
print("Skewness = ")
print(skewnesses)
print()
print("Kurtosis = ")
print(kurtoses)
print()
```

```python
# Plot Mean
plt.figure(figsize=(7, 5))
plt.plot(means,color='red',marker="o")
plt.title('Mean of Each Column')
plt.xlabel('Column Index')
plt.ylabel('Mean')
plt.tight_layout()
plt.show()
```

```python
# Plot Variance
plt.figure(figsize=(7, 5))
plt.plot(variances,color='orange',marker="o")
plt.title('Variance of Each Column')
plt.xlabel('Column Index')
plt.ylabel('Variance')
plt.tight_layout()
plt.show()
```

```python
# Plot Skewness
plt.figure(figsize=(7, 5))
plt.plot(skewnesses,color='green',marker="o")
plt.title('Skewness of Each Column')
plt.xlabel('Column Index')
plt.ylabel('Skewness')
plt.tight_layout()
plt.show()
```

```python
# Plot Skewness
plt.figure(figsize=(7, 5))
plt.plot(skewnesses,color='black',marker="o")
plt.title('Skewness of Each Column')
plt.xlabel('Column Index')
plt.ylabel('Skewness')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist(means, bins=10, edgecolor='black')
plt.title('Histogram of Means')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist(variances, bins=10, edgecolor='black')
plt.title('Histogram of Varience')
plt.xlabel('Varience')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist(kurtoses, bins=10, edgecolor='black')
plt.title('Histogram of Kurtosis')
plt.xlabel('Kurtosis')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist(skewnesses, bins=10, edgecolor='black')
plt.title('Histogram of skewnesses')
plt.xlabel('skewnesses')
plt.ylabel('Frequency')
plt.show()
```
