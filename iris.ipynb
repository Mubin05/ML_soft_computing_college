```python
from ucimlrepo import fetch_ucirepo 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets
print(X.describe())
```

```python
iris_arr = X.to_numpy()
column_names = X.columns
# Calculate the mean and variance of each column
mean = np.mean(iris_arr, axis=0)
variance = np.var(iris_arr, axis=0)
# Print the mean and variance of each column
for col, m, v in zip(column_names, mean, variance):  # zip helps to iterate through more than one array at a time
    print(f"Mean of {col} : {m:.3f}")
    print(f"Variance of {col} : {v:.3f} \n")
```

```python
# Plot the mean of each column
plt.figure(figsize=(10, 6))
plt.bar(column_names, mean, color='skyblue')
plt.title('Mean of Each Feature')
plt.xlabel('Features')
plt.ylabel('Mean')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Plot the variance of each column
plt.figure(figsize=(10, 6))
plt.bar(column_names, variance, color='lightgreen')
plt.title('Variance of Each Feature')
plt.xlabel('Features')
plt.ylabel('Variance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the mean and variance for each column
x = np.arange(len(column_names))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for the mean
mean_bars = ax.bar(x - width/2, mean, width, label='Mean', color='skyblue')

# Plot bars for the variance
var_bars = ax.bar(x + width/2, variance, width, label='Variance', color='lightgreen')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Features')
ax.set_ylabel('Values')
ax.set_title('Mean and Variance of Each Feature')
ax.set_xticks(x)
ax.set_xticklabels(column_names, rotation=45)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
